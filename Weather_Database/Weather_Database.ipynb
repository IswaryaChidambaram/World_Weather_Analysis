```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from citipy import citipy
import requests
import json
from config import weather_api_key
```

```python
lats=np.random.uniform(-90,90,size=2000)
lngs=np.random.uniform(-90,90,size=2000)
lat_lngs=zip(lats,lngs)
lat_lngs
```

```python
coordinates = list(lat_lngs)
coordinates
```

```python
cities = []

# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
# Print the city count to confirm sufficient count.

    print(city)
print(len(cities))
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
print(url)
```

```python
city="Mataura"
query_url = url + "appid=" + weather_api_key + "&q=" + city + "&units=metric"
weather_response = requests.get(query_url)
weather_json = weather_response.json()
print(json.dumps(weather_json))

```

```python
weather_data=[]

for i,city in enumerate(cities):
    
    print("Retrieving data for " +  city + " : city "+ str(i))

    query_url = url + "appid=" + weather_api_key + "&q=" + city + "&units=metric"
    try:
        weather_response = requests.get(query_url)
        weather_json = weather_response.json()


        weather={}
        weather["city"]=weather_json["name"]
        weather["country"]=weather_json["sys"]["country"]
        weather["latitude"]=weather_json["coord"]["lat"]
        weather["longitude"]=weather_json["coord"]["lon"]
        weather["Maximum_temperature"]=weather_json["main"]["temp_max"]
        weather["percent_humidity"]=weather_json["main"]["humidity"]
        weather["percent_cloudiness"]=weather_json["clouds"]["all"]
        weather["wind_speed"]=weather_json["wind"]["speed"]
        weather["description"]=weather_json["weather"][0]["description"]
        print(weather)
        weather_data.append(weather) 
        
    except Exception as e:
        print("Some error occured while retrieving data for " + city+ ": " )
        print(e)
      
    

    
```

```python
weather_data

```

```python
city_data_df=pd.DataFrame(weather_data)
city_data_df
```

```python
output_data_file = "WeatherPy_Database.csv"
city_data_df.to_csv(output_data_file, index_label="City_ID")
```
